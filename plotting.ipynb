```python
import xarray as xr
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Load the .cdf file
file_path = 'data/3.cdf'
dataset = xr.open_dataset(file_path)

# Extract temperature data and time
temperature = dataset['temp'].values  # Assuming temperature has shape (time, x, y, z)
time = dataset['Time'].values         # Assuming time has length matching the first dimension of 'temp'

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Initialize the plot with the first frame of data (assuming z=0, y=0 slice)
cax = ax.imshow(temperature[0, :, 0, :], cmap='viridis')
fig.colorbar(cax, label='Temperature')

ax.set_title(f'Temperature Data at Time {time[0]:.4f}')
ax.set_xlabel('X')
ax.set_ylabel('Z')

# Update function for animation
def update(frame):
    # Print to check the current frame
    print(f"Frame: {frame}, Time: {time[frame]:.4f}")

    # Update the data for the plot for the current frame
    cax.set_data(temperature[frame, :, 0, :])  # Slice to get the (x, z) plane at y=0
    ax.set_title(f'Temperature Data at Time {time[frame]:.4f}')

    return cax,

# Create the animation and assign it to a variable 'ani'
print(len(time))
ani = FuncAnimation(fig, update, frames=10, interval=100)

# Show the animation
plt.show()
```
